```python
%matplotlib inline

import numpy as np
from sklearn.datasets import load_sample_image
import seaborn
```

```python
img=load_sample_image('china.jpg')
```

```python
img.shape
```

```python
x=img
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.grid(False)
```

```python
x=x.reshape(-1,3)
```

```python
x
```

```python
x=x/255.0
```

```python
x
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
mk=MiniBatchKMeans(3)
```

```python
mk
```

```python
label=mk.fit_predict(x)
```

```python
label
```

```python
label.shape
```

```python
color=mk.cluster_centers_
```

```python
color
```

```python
new_img=(color[label].reshape(img.shape)*255).astype(np.uint8)
```

```python
img
```

```python
new_img
```

```python
new_img
```

```python
plt.imshow(new_img)
plt.grid(False)
```
